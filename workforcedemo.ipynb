```python
from camel.agents.chat_agent import ChatAgent
from camel.messages.base import BaseMessage
from camel.models import ModelFactory
from camel.societies.workforce import Workforce
from camel.tasks.task import Task
from camel.configs import DeepSeekConfig,ChatGPTConfig
from camel.types import ModelPlatformType, ModelType
import os

# model = ModelFactory.create(
#     model_platform=ModelPlatformType.OPENAI_COMPATIBLE_MODEL,
#     model_type="deepseek-chat",
#     api_key=os.environ["DEEPSEEK_API_KEY"],
#     url="https://api.openai.com/v1",
#     model_config_dict={"temperature": 0.4},)

# model = ModelFactory.create(
#     model_platform=ModelPlatformType.DEEPSEEK,
#     model_type=ModelType.DEEPSEEK_CHAT,
#     model_config_dict=DeepSeekConfig(temperature=0.2).as_dict(),
# )

model = ModelFactory.create(
    model_platform=ModelPlatformType.OPENAI,
    model_type=ModelType.GPT_4O_MINI,
    model_config_dict=ChatGPTConfig().as_dict(),
)

teacher_agent = ChatAgent(
    BaseMessage.make_assistant_message(
        role_name="雅思口语老师",
        content="你是雅思口语老师，请根据学生的需求，给出相应的建议和指导",
    ),
    model=model,
)

# Set up traveler agent
student_agent = ChatAgent(
    BaseMessage.make_assistant_message(
        role_name="雅思考试学生",
        content="你是雅思考试学生，请根据雅思口语老师的建议，给出相应的回答",
    ),
    model=model,
)

workforce = Workforce(
    'A travel group',
    coordinator_agent_kwargs = {"model": model},
    task_agent_kwargs = {"model": model},
)

workforce.add_single_agent_worker(
    "A teacher",
    worker=teacher_agent,
).add_single_agent_worker(
    "A student", worker=student_agent
)
# specify the task to be solved
human_task = Task(
    content=(
        "雅思口语考试"
    ),
    id='0',
)

task = workforce.process_task(human_task)

print('Final Result of Original task:\n', task.result)
```
